```python
import pandas
import numpy as np
import seaborn as sb
from sklearn import datasets
import matplotlib.pyplot as plot

plot.style.use('ggplot')
%matplotlib inline

iris = datasets.load_iris()
iris_df_dat = pandas.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df_tg = pandas.DataFrame(iris['target'],columns=['species'])
iris_dat_all = pandas.concat([iris_df_dat,iris_df_tg],axis=1)
```

```python
iris['feature_names']
```

```python
sb.pairplot(iris_dat_all,hue='species')
```

```python
irisd = iris_dat_all
fig,ax = plot.subplots(2,2,figsize=(10,10))
sb.set(style='white', palette='muted')

sb.violinplot(x=irisd['species'], y=irisd['sepal length (cm)'], ax=ax[0,0])
sb.violinplot(x=irisd['species'], y=irisd['sepal width (cm)'], ax=ax[0,1])
sb.violinplot(x=irisd['species'], y=irisd['petal length (cm)'], ax=ax[1,0])
sb.violinplot(x=irisd['species'], y=irisd['petal width (cm)'], ax=ax[1,1])

fig.suptitle('Violin Plots', fontsize=18)
fig.tight_layout()
```
