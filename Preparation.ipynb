```python
import pandas
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plot
plot.style.use('ggplot')
%matplotlib inline
```

```python
iris_d = datasets.load_iris()
```

```python
iris_d.keys()
```

```python
iris_d['target_names']
```

```python
iris_df_dat = pandas.DataFrame(iris_d['data'], columns=iris_d['feature_names'])
```

```python
iris_df_dat
```

```python
iris_df_tg = pandas.DataFrame(iris_d['target'],columns=['species'])
```

```python
iris_d = pandas.concat([iris_df_dat,iris_df_tg],axis=1)
```

```python
iris_d['species'] = iris_d['species'].map({0:'Setosa',1:'Versicolor',2:'Virginica'})
iris_d
```

```python
iris_d['wide petal'] = iris_d['petal width (cm)'].apply(lambda v: 1 if v >= 1.3 else 0)
iris_d['petal area'] = iris_d.apply(lambda v: v['petal width (cm)']*v['petal length (cm)'], axis=1)
iris_d
```

```python
iris_d.groupby('species').mean()
```

```python
iris_d.groupby(lambda x: x<10)['species'].unique().apply(lambda v: 1 if len(v)==3 else 0).to_frame()
```
